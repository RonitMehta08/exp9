```python
#declaration of numpy
import numpy as np

a = np.array([1,10,20,30,40,50,500])
b = np.array([[1,2,3],[4,5,6]])

print(a)
print(b)
```

```python
#dimension of array

print(a.ndim)
print(b.ndim)
```

```python
#shape of matrix
print(a.shape)
print(b.shape)
```

```python
#printing datatype of a
print(a.dtype)
print(b.dtype)
```

```python
print(np.zeros((2,3)))
print(np.ones((2,3)))
```

```python
print(np.eye((3)))
```
